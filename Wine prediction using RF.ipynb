```python
# Now, next step is data pre-processing. 
# Means separate the label column (quality) and data set columns (all 11 parameters of the data set)
```

```python
# Put the data set into X variable.

X=wine_data.drop('quality',axis=1)
```

```python
print(X)
```

```python
# Label data store into Y variable but before that label (quality column) is binarization (means values converted into 0 & 1) 
# its make easy to process on the data.
# If the label value 7 or greater than 7 that is replaced with 1 and represents good quality of wine.
# if the label value less than 7 (3,4,5,6) than that is replaced with 0 and represents bad qaulity of wine. 
```

```python
Y=wine_data['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
# Now Y variable carry the label of the data in the form of 0 & 1.

print(Y)
```

```python
# Spliting the data set into traingn and testing sets. 

x_train,x_test,y_train,y_test= train_test_split(X,Y,test_size=0.15,random_state=3)
```

```python
print(Y.shape,y_train.shape,y_test.shape)
```

```python
# Train the model using Random Forest. 

model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
# Check the accuracy of the mmodel on the training data set.

x_train_prediction=model.predict(x_train)
```

```python
x_train_accuracy=accuracy_score(x_train_prediction,y_train)
print("accuracy of the model on the training set", x_train_accuracy)
```

```python
# Now the prediction of the model on the testing data set. 

x_test_prediction = model.predict(x_test)
```

```python
x_test_accuracy = accuracy_score(x_test_prediction,y_test)
print("Accuracy of the model on the testing data",x_test_accuracy)
```
